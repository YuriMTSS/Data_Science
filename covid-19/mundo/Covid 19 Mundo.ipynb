```python
import pandas as pd
import numpy as np
```

# Casos confirmados de COVID no mundo

```python
confirmados = pd.read_csv('cases_covid_confirmed_world.csv')
confirmados
```

```python
confirmados.shape
```

```python
confirmados.iloc[:, 4: confirmados.shape[1]]
```

```python
confirmados = np.sum(confirmados.iloc[: ,4: confirmados.shape[1]])
confirmados.index = pd.DatetimeIndex(confirmados.index)
confirmados.head()
```

```python
confirmados.plot()
```

# Mortes por COVID no mundo

```python
mortes = pd.read_csv('case_covid_deaths_world.csv')
mortes
```

```python
mortes.shape
```

```python
mortes = np.sum(mortes.iloc[:, 4: mortes.shape[1]])
mortes.index = pd.DatetimeIndex(mortes.index)
mortes.head()
```

```python
mortes.plot()
```

# **MORTALIDADE**

```python
mortalidade = (mortes/confirmados) * 100
mortalidade.head()
```

```python
mortalidade.plot()
```

# Pacientes recuperados

```python
recuperados = pd.read_csv('recovery.csv')
recuperados
```

```python
recuperados.shape
```

```python
recuperados = np.sum(recuperados.iloc[:, 4 : recuperados.shape[1]])
recuperados.index = pd.DatetimeIndex(recuperados.index)
recuperados.head()
```

```python
recuperados.plot()
```

```python
ativos = confirmados - (recuperados + mortes)
ativos.head()
```

```python
ativos.plot()
```

```python
df = pd.concat([confirmados, mortes, ativos, recuperados], axis = 1)

df.columns = (['confirmados','mortes','ativos','recuperados'])

df.head()
```

```python
df.plot()
```
