```python
name_type = pokemon[['english_name','primary_type','secondary_type']]
name_type.head()
```

```python
filtro_pokemon = pokemon[pokemon['primary_type'] == 'grass']
filtro_pokemon.tail()
```

```python
pokemon_fraquezas = pokemon[pokemon['against_water'] > 2]
pokemon_fraquezas[['english_name', 'primary_type', 'secondary_type','against_water']]
```

```python
pokemon_dupla_fraqueza_E = pokemon[(pokemon['against_dark'] > 1) & (pokemon['against_ghost'] > 1)]
pokemon_dupla_fraqueza_E[['english_name', 'primary_type', 'secondary_type', 'against_dark', 'against_ghost']].head(20)
```

```python
pokemon_dupla_fraqueza_OU = pokemon[(pokemon['against_water'] > 1) | (pokemon['against_ground'] > 1)]
pokemon_dupla_fraqueza_OU[['english_name', 'primary_type', 'secondary_type', 'against_water', 'against_ground']].head(20)
```

```python
pokemon_dupla_fraqueza_OU = pokemon[(pokemon['against_water'] > 1) & (pokemon['against_ground'] > 1) & (pokemon['against_electric'])]
pokemon_dupla_fraqueza_OU[['english_name', 'primary_type', 'secondary_type', 'against_water', 'against_ground', 'against_electric']].head(20)
```
