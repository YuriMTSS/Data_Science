```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

# Documentação Seaborn
https://seaborn.pydata.org/examples/index.html#

```python
iris = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
iris
```

```python
sns.scatterplot(x = 'sepal.length',y = 'sepal.width', data = iris)
plt.title('Sepal length VS Sepal Width')
```

```python
sns.regplot(x = 'sepal.length', y = 'sepal.width', data = iris)
plt.title('Sepal Length vs. Sepal Width com Regression Fit')
```

```python
sns.scatterplot(x = 'sepal.length', y = 'sepal.width', data = iris, marker = '+')
plt.title('Sepal Length vs. Sepal Width')
```

```python
f, ax = plt.subplots(figsize=(15, 5))
sns.despine(f)
sns.kdeplot(iris['petal.length'])
```

```python
f, ax = plt.subplots(figsize=(15, 5))
sns.despine(f)
sns.kdeplot(iris['petal.length'], shade=True)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.kdeplot(iris['petal.width'], vertical = True)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.kdeplot(iris['petal.width'], vertical=True, shade=True)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.kdeplot(iris['petal.width'], shade=True)
sns.kdeplot(iris['petal.length'], shade=True)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.kdeplot(iris['sepal.length'], shade=True, color='r')
sns.kdeplot(iris['sepal.width'], shade=True, color='m')
```

```python
sns.kdeplot(iris['sepal.length'], iris['sepal.width'])
plt.title("Sepal Length vs Sepal Width 2D Density Plot")
```

```python
sns.kdeplot(iris['sepal.length'], iris['sepal.width'], cmap="Reds", shade=True, bw=.15)
```

```python
sns.kdeplot(iris['sepal.length'], iris['sepal.width'], cmap="Blues", shade=True, shade_lowest=True, )
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.barplot(x='variety', y='sepal.length', data=iris, palette='magma')
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.barplot(x='sepal.length', y='variety', data=iris, palette='magma', orient='h')
```

```python
sns.violinplot(x=iris['sepal.length'])
```

```python
sns.violinplot(y=iris['sepal.length'], x=iris['sepal.width'])
```
